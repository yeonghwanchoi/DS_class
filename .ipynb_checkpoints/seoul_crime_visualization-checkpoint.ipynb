### 1. how often have the viorent crimes occurred in each area?

```python
# index is consisting of stations
crime_station.index
```

```python
#not all the stations represent the area where these are located in
#match the stations with its areas  
station_name=["Hyehwa","namdaemoon","Jungbu","Jongam","Seobu","Bangbae","Suseo"]
gu_name = ["jongno","Jungbu","Jungbu","Seongbuk","Eunpyeong","Seocho","Gangnam"]
```

```python
tmp_gu=[]
for idx, _ in crime_station.iterrows():
    if idx in station_name:
        tmp_gu.append(gu_name[station_name.index(idx)])
    else:
        tmp_gu.append(idx)
```

```python
#add new area column
crime_station["area"]= tmp_gu
crime_station.head(2)
```

```python
#new dataframe sorted by area
crime_area = pd.pivot_table(crime_station, index="area", aggfunc=np.sum)
crime_area.head()
```

```python
#crime arrest rates columns on dataframe
arrest=["rape_arrest","burglar_arrest","merder_arrest","theft_arrest","violence_arrest"]
occurs=["rape_incident","burglar_incident","merder_incident","theft_incident","violence_incident"]
new_columns=["rape_arrest_rate","burglar_arrest_rate","merder_arrest_rate","theft_arrest_rate","violence_arrest_rate"]
crime_area[new_columns]=crime_area[occurs]/(crime_area[arrest].values)*100
crime_area.head()
```

```python
crime_area= crime_area.drop(arrest,axis=1)
crime_area[crime_area[new_columns]>100]=100
```

```python
#rename columns 
crime_area.rename(columns={
    "rape_incident" : "rape",
    "theft_incident" : "theft",
    "merder_incident" : "merder",
    "burglar_incident" : "burglar",
    "violence_incident" : "violence"},
    inplace = True)
```

```python
crime_area.head()
```

```python
#normalization 
col=["merder","burglar","rape","theft","violence"]
col2= ["merder_arrest_rate","burglar_arrest_rate","rape_arrest_rate","theft_arrest_rate","violence_arrest_rate"]

crime_area_norm = crime_area[col] / crime_area[col].max()
crime_area_norm[col2]= crime_area[col2]
crime_area_norm.head()
```

```python
crime_area_norm["crime"] = np.mean(crime_area_norm[col],axis=1)
crime_area_norm.head()
```

```python
crime_area_norm["arrest"] = np.mean(crime_area_norm[col2],axis=1)
crime_area_norm.head()
```
